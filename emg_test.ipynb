```python
#reads samples from serial channel, expects it to be sent as purely tab-deliminated
def readSamples(filename,socket=serial,chnum=channel_num,snum=sample_num):
    with open('%s.csv'%filename,'a') as f:
        #Title row of csv
        #f.write('\t'.join(('CH%d'%x for x in range(chnum))))
        #f.write('\r\n')
        iters = 0
        socket.reset_input_buffer()
        while iters < snum:
            inp = socket.read_until()
            try:
                out = inp.decode('utf8')
                if out[0].isdigit():
                    f.write(out)
                    iters += 1
                    if iters % int(snum/4) == 0: print("%d samples"%iters)
                else:
                    print("INFO: %s"%out)
            except:
                pass
        print("DONE!")
```
